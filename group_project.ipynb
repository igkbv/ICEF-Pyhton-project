```python
import requests               # for sending HTTP requests to websites
import time                   # for measuring response time
from datetime import datetime # for recording the date and time of each check
```

```python
# Create a function to check a website
def check_website(url, timeout=5):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    start_time = time.time()

    try:
        headers = {"User-Agent": "Mozilla/5.0"}
        response = requests.get(url, headers=headers, timeout=5)
        end_time = time.time()

        result = {
            "url": url,
            "timestamp": timestamp,
            "status": "success",
            "status_code": response.status_code,
            "response_time": round(end_time - start_time, 3),
            "error_type": None
        }
    
    # If the server does not respond within the allowed time
    except requests.exceptions.Timeout:
        result = {
            "url": url,
            "timestamp": timestamp,
            "status": "failed",
            "status_code": None,
            "response_time": None,
            "error_type": "timeout"
        }

    # If the connection to the website cannot be established
    except requests.exceptions.ConnectionError:
        result = {
            "url": url,
            "timestamp": timestamp,
            "status": "failed",
            "status_code": None,
            "response_time": None,
            "error_type": "connection_error"
        }

    # If any other request-related error occurs
    except requests.exceptions.RequestException:
        result = {
            "url": url,
            "timestamp": timestamp,
            "status": "failed",
            "status_code": None,
            "response_time": None,
            "error_type": "request_error"
        }

    return result
```
